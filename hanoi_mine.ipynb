```python
# Recursive program for Tower of Hanoi 
  
# Recursive function to solve Tower  
# of Hanoi puzzle 
it=0
def towerOfHanoi(n, from_rod, to_rod, aux_rod1, 
                                     aux_rod2): 
  
    if (n == 0): 
        return
    if (n == 1): 
        print("Move disk", n, "from rod",  
                 from_rod, "c to rod",  to_rod) 
        return
      
  
    towerOfHanoi(n - 2, from_rod, aux_rod1, 
                              aux_rod2, to_rod) 
    print("Move disk", n-1, "from rod", from_rod, 
                           "c to rod", aux_rod2) 
                  
    print("Move disk", n, "from rod", from_rod, 
                             "c to rod", to_rod) 
                  
    print("Move disk", n-1, "from rod", aux_rod2, 
                             "c to rod", to_rod) 
              
    towerOfHanoi(n - 2, aux_rod1, to_rod,  
                              from_rod, aux_rod2)
```

```python
##RANDOM INITIAL STATE BUT ONLY FOR 3 PIGS

iteration=0
def optimal_move(initial_state, biggest_disk, final_peg):
    for wrong_position in range(biggest_disk, len(initial_state)):
        #Defining the current position of the disk
        current_peg = initial_state[wrong_position]         
        if current_peg != final_peg:

            #First, we need to find the biggest disk positioned in a wrong peg

            #Finding the possible movement
            possible_move = 3 - final_peg - current_peg

            #First, we need to move the smaller disk who are over the targeted disk
            optimal_move(initial_state, wrong_position+1, possible_move)


            print (wrong_position, ":", current_peg,   " --> ", final_peg)
            initial_state[wrong_position]=final_peg
            global iteration
            iteration=iteration+1

            #Final move of the smaller disks
            optimal_move(initial_state, wrong_position+1, final_peg)
            
            break;
    
    return(iteration)
```
